# Stock Portfolio Analysis
### The Data
I will be using stock data obtained from Yahoo from 01.01.2015 till 01.01.2020.
### The Goal
Evaluate the portfolio of four stocks (Apple, Cisco, Tesla, Amazon) devided into equal parts and find the best combination of stocks using Sharpe ratio.

```python
import numpy as np
import pandas_datareader.data as wb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
assets = ['AAPL','CSCO','TSLA','AMZN']
df = pd.DataFrame()

for x in assets:
    df[x] = wb.DataReader(x, data_source='yahoo', start='2015-1-1',end = '2020-01-01')['Adj Close']
```

```python
num_ports = 6000
pf_returns = np.zeros(num_ports)
pf_volatilities = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)
all_weights = np.zeros((num_ports, len(assets)))
```

```python
for x in range(num_ports):
    weights=np.array(np.random.random(len(assets)))
    weights/=np.sum(weights)
    all_weights[x,:] = weights
    pf_returns[x] = np.sum(log_returns.mean()*weights*252)
    pf_volatilities[x] = np.sqrt(np.dot(weights.T,np.dot(log_returns.cov()*252,weights)))
    sharpe_arr[x] = (pf_returns[x]-0.025)/pf_volatilities[x]
```

```python
pf_returns=np.array(pf_returns)
pf_volatilities=np.array(pf_volatilities)
```

```python
print('Max sharpe ratio in array: {}'.format(sharpe_arr.max()))
print('Its location in array: {}'.format(sharpe_arr.argmax()))
```

```python
print(all_weights[sharpe_arr.argmax()])
```

```python
max_pf_returns = pf_returns[sharpe_arr.argmax()]
max_pf_volatilities = pf_volatilities[sharpe_arr.argmax()]
```

```python
plt.figure(figsize=(12,8))
plt.scatter(pf_volatilities, pf_returns, c=sharpe_arr, cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.scatter(max_pf_volatilities, max_pf_returns,c='red', s=50) # red dot
plt.show()
```

Expected Portfolio Variance:

```python
port_var=np.dot(all_weights[sharpe_arr.argmax()].T,np.dot(log_returns.cov()*252,all_weights[sharpe_arr.argmax()]))
port_var
```
